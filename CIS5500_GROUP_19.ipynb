# Dataset Tourist Hotspots

```python
hotspot = pd.read_csv('/content/drive/MyDrive/cis550data/cleaned_data_USA.csv')
hotspot.head(10)
```

```python
hotspot = hotspot.rename(columns= {'Unnamed: 0':'index'})
hotspot.head()
```

```python
hotspot.describe()
```

```python
#drop columns that has null value and the columns that will not be used in our web platform
hotspot.drop(['zipcode', 'Weighted_Score', 'Weighted_Average', 'All_Cities'], axis = 1, inplace = True)
hotspot.head()
```

```python
#check for null values
hotspot.isnull().values.any()
```

```python
#check duplicates
duplicates = hotspot[hotspot.duplicated(subset=['name', 'address'], keep=False)]
print(duplicates)
```

```python
hotspot = hotspot.drop_duplicates(subset = ['name','address'])
```

```python
hotspot.duplicated(subset= (['name', 'address'])).any()
```

```python
hotspot_unique_cities = hotspot['city'].unique()
print(hotspot_unique_cities)
print(hotspot['city'].nunique())
```

```python
hotspot['city'] = hotspot['city'].str.replace('-', ' ').str.title()
hotspot['city'].unique()
```

```python
hotspot['city'] = hotspot['city'].str.replace('Washington D C', 'Washington D.C.')
```

```python
hotspot['city'].unique()
```

```python
hotspot['state'].unique()
```

```python
#By inspecting the raw data, we found the state of Washington D.C. is wrong, change the state NE to DC for future use
hotspot.loc[hotspot['city']=='Washington D.C.', 'state'] = 'DC'
```

```python
hotspot.loc[hotspot['city'] == 'St Louis', 'city'] = 'St. Louis'
```

```python
hotspot [hotspot['state'] == 'DC']
```

```python
hotspot[hotspot['city'] == 'St. Louis']
```

```python
hotspot.dtypes
```

#City

```python
uscities = pd.read_csv('/content/drive/MyDrive/cis550data/uscities.csv')
uscities.head()
```

```python
#change washington in state DC to Washington D.C.
uscities.loc[(uscities['city'] == 'Washington') & (uscities['state_id'] == 'DC'), 'city'] = 'Washington D.C.'
```

```python
uscities.loc[(uscities['city'] == 'New York') & (uscities['state_id'] == 'NY'), 'city'] = 'New York City'
```

```python
#rename state_id to state for later merging
uscities = uscities.rename(columns={'state_id': 'state', 'lat': 'latitude', 'lng':'longitude'})
```

```python
merged_data = pd.merge(hotspot, uscities, on=['city', 'state'], how='inner')
merged_data.head()
```

```python
city = merged_data[['city','state','longitude','latitude']].drop_duplicates()
city.head()
```

```python
print(len(city))
```

```python
duplicated_cities = city[city['city'].duplicated(keep=False)]
print(duplicated_cities)
```

```python
print(city['city'].nunique())
```

```python
print(city['city'].unique())
```

```python
print(hotspot['city'].unique())
```
